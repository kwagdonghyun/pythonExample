```python
a = 200
b = 33
c = 500

if a > b and c > a:
    print("Both conditions are True")

if a > b or a > c:
    print("One condition is True")

if not a < b:
    print("a is not less than b")

# Nested If
age = 25
has_license = True

if age >= 18:
    if has_license:
        print("You can drive")
    else:
        print("Need a license")
else:
    print("Too young")
```

```python
day = 3

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case _:
        print("Other")
```
