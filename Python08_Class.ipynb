```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def birthday(self):
        self.age += 1

p1 = Person("Kim", 20)

p1.birthday()

print(p1.age)
```

```python
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname

    def printname(self):
        print(self.firstname, self.lastname)

class Student(Person):
    pass

x = Student("Kim", "Minsu")
x.printname()
```

```python
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname

class Student(Person):
    def __init__(self, fname, lname, year):
        super().__init__(fname, lname)
        self.year = year

x = Student("Kim", "Minsu", 2025)

print(x.firstname)
print(x.year)
```

```python
class Car:
    def move(self):
        print("Drive!")

class Boat:
    def move(self):
        print("Sail!")

class Plane:
    def move(self):
        print("Fly!")

for x in (Car(), Boat(), Plane()):
    x.move()
```

```python
class Person:
    def __init__(self, age):
        self.__age = age

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age

p = Person(20)

print(p.get_age())

p.set_age(25)

print(p.get_age())
```
