```python
cars = ["Ford", "Volvo", "BMW"]

print(cars)
print(cars[0])

cars.append("Honda")
print(cars)

cars.pop(1)
print(cars)

cars.remove("BMW")
print(cars)
```

```python
import datetime

today = datetime.datetime.now()

print(today)
print(today.year)
print(today.strftime("%A"))

birthday = datetime.datetime(2024, 5, 17)
print(birthday)
```

```python
import math

print(min(5, 10, 25))
print(max(5, 10, 25))
print(abs(-7.25))
print(pow(4, 3))

print(math.sqrt(64))
print(math.ceil(1.4))
print(math.floor(1.4))
print(math.pi)
```

```python
import json

person = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

json_data = json.dumps(person)
print(json_data)

python_data = json.loads(json_data)
print(python_data["name"])

print(json.dumps(person, indent=4))
```
