```python
def greet():
    print("Hello")

greet()
```

```python
def greet(name):
    print("Hello", name)

greet("Kim")
```

```python
def add(a, b):
    return a + b

print(add(3, 5))
```

```python
def country(name, nation="Korea"):
    print(name, nation)

country("Kim")
country("John", "USA")
```

```python
def total(*numbers):
    print(sum(numbers))

total(1, 2, 3, 4)
```

```python
def student(**info):
    for key, value in info.items():
        print(key, value)

student(name="Kim", age=23)
```

```python
x = 100

def test():
    x = 50
    print(x)

test()
print(x)
```

```python
square = lambda x: x * x
print(square(5))

multiply = lambda a, b: a * b
print(multiply(3, 4))
```

```python
numbers = [1, 2, 3, 4, 5]
double = list(map(lambda x: x * 2, numbers))
print(double)
```

```python
odd = list(filter(lambda x: x % 2 == 1, numbers))
print(odd)
```

```python
students = [("Kim", 23), ("Lee", 20), ("Park", 25)]
print(sorted(students, key=lambda x: x[1]))
```

```python
def countdown(n):
    if n == 0:
        print("Done")
    else:
        print(n)
        countdown(n - 1)

countdown(5)
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(7))
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3

for value in my_generator():
    print(value)
```

```python
def simple_gen():
    yield "A"
    yield "B"

gen = simple_gen()

print(next(gen))
print(next(gen))
```
