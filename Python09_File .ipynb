```python
with open("demofile.txt", "w") as f:
    f.write("Hello! Welcome to demofile.txt\n")
    f.write("This file is for testing purposes.")

f = open("demofile.txt", "r")
print(f.read())
f.close()
```

```python
with open("demofile.txt", "w") as f:
    f.write("Python File Example")

with open("demofile.txt", "r") as f:
    print(f.read())
```

```python
with open("demofile.txt", "w") as f:
    f.write("First Line\nSecond Line")

f = open("demofile.txt", "r")
print(f.readline())
f.close()
```

```python
with open("demofile.txt", "w") as f:
    f.write("Hello Python")

with open("demofile.txt", "r") as f:
    print(f.read(5))
```

```python
with open("demofile.txt", "w") as f:
    f.write("Apple\nBanana\nOrange")

with open("demofile.txt", "r") as f:
    print(f.readline())
    print(f.readline())
```

```python
with open("demofile.txt", "w") as f:
    f.write("Apple\nBanana\nOrange")

with open("demofile.txt", "r") as f:
    for line in f:
        print(line.strip())
```
